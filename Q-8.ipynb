Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import matplotlib.pyplot as pltactor_kill_counts
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data8 = pd.read_csv(r'C:\Users\DELL\Desktop\Monish\Data Programming\Assignment-3\actor_kill_counts.csv')

objects = data8['Actor']
y_pos = np.arange(len(objects))
performance = data8['Count']

plt.bar(y_pos, performance, align='center', alpha=1, width=0.5)
plt.xticks(y_pos,objects)
plt.ylabel('Number of Kills')
plt.title('Actors with their kill count')

plt.show()
```
