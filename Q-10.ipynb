## Question 10
Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.

Don't forget to label your axes!

Color each dot according to its year.

```python
import pandas as pd
import matplotlib.pyplot as plt

data10 = pd.read_csv(r'C:\Users\DELL\Desktop\Monish\Data Programming\Assignment-3\arcade-revenue-vs-cs-doctorates.csv')


colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}


fig, ax = plt.subplots()

for i in range(len(data10['Year'])):
    ax.scatter(data10['Total Arcade Revenue (billions)'][i], data10['Computer Science Doctorates Awarded (US)'][i])

ax.set_title('Relationship between Arcade Revenue and Computer Science Doctorates Awarded in USA from 2000 to 2009  ')
ax.set_xlabel('Total Arcade Revenue (billions)')
ax.set_ylabel('Computer Science Doctorates Awarded (US)')
```
