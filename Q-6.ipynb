## Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv(r'C:\Users\DELL\Desktop\Monish\Data Programming\Assignment-3\us-marriages-divorces-1867-2014.csv')

# get columns to plot
columns = data.columns
# create x data
x_data = range(0, data.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column

ax.plot(x_data, data['Marriages'],label='Marriages')
ax.plot(x_data, data['Divorces_per_1000'],label='Divorces per capita ')
# set title and legend
ax.set_title('Line Graph of Marriages and Divorces per capita')
ax.legend()

```
